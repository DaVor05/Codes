```python
from pyspark.sql import functions as F

# Datos de ejemplo
ventas = [
    ("2025-01-01", "Producto A", 10, 100.0),
    ("2025-01-01", "Producto B", 5, 50.0),
    ("2025-01-02", "Producto A", 7, 70.0),
    ("2025-01-02", "Producto B", 3, 30.0),
    ("2025-01-02", "Producto C", 2, 40.0)
]

columnas = ["Fecha", "Producto", "Cantidad", "Total"]

# Crear DataFrame
df_ventas = spark.createDataFrame(ventas, columnas)

# Mostrar datos originales
df_ventas.show()

# Calcular ventas totales por producto
df_ventas.groupBy("Producto").agg(
    F.sum("Cantidad").alias("Unidades_Vendidas"),
    F.sum("Total").alias("Ingresos_Totales"),
    F.avg("Total").alias("Promedio_Venta")
).show()

# Calcular ventas totales por d√≠a
df_ventas.groupBy("Fecha").agg(
    F.sum("Total").alias("Ingresos_Diarios")
).orderBy("Fecha").show()
```
